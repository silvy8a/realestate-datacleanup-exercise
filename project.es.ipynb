#### Exercise 19. What can you say about the price per square meter (price/m2) between the towns of "Getafe" and "Alcorcón"? You must use the subset obtained in the question 14 (★★☆)

Hint: Create a new column called `pps` (price per square) and then analyse the values

```python
import pandas as pd
import matplotlib.pyplot as plt
ds = pd.read_csv('assets/real_estate.csv', sep=';')
cinturon_sur=ds.apply(lambda row: row[ds['level5'].isin(['Fuenlabrada','Leganés', 'Getafe', 'Alcorcón'])])
cinturon_sur['pps'] = cinturon_sur["price"] / cinturon_sur["surface"]
alcorcon_pps=cinturon_sur[cinturon_sur['level5'] == "Alcorcón"]['pps'].mean()
getafe_pps=cinturon_sur[cinturon_sur['level5'] == "Getafe"]['pps'].mean()
print(round(alcorcon_pps,2))
print(round(getafe_pps,2))

```

Según los cálculos, el precio por m2 es superior en Alcorcón (2239,30€) frente a Getafe (2066,31€). Teniendo en cuenta que hemos visto anteriormente que el promedio del precio de la vivienda es superior en Getafe, podríamos deducir que hay menos superficie disponible en Alcorcón.
